```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data for the Sangvi dataset
sangvi = pd.DataFrame(np.random.randn(100, 4), columns=["CtN", "CtE", "CtRdRp", "PCR_Result"])
sangvi["PCR_Result"] = np.where(sangvi["PCR_Result"] > 0, "Positive", "Negative")
sangvi.to_csv("sangvi.csv", index=False)

# Generate random data for the Baner dataset
baner = pd.DataFrame(np.random.randn(100, 4), columns=["CtN", "CtE", "CtRdRp", "PCR_Result"])
baner["PCR_Result"] = np.where(baner["PCR_Result"] > 0, "Positive", "Negative")
baner.to_csv("baner.csv", index=False)

# Load the datasets from the CSV files
sangvi = pd.read_csv("sangvi.csv")
baner = pd.read_csv("baner.csv")

# Calculate the correlation matrix between the two datasets
corr = pd.concat([sangvi, baner]).corr()

# Plot the correlation heatmap
sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap="coolwarm")
plt.title("Correlation Heatmap between Sangvi and Baner Datasets")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data for the Sangvi dataset
sangvi = pd.DataFrame(np.random.randn(100, 4), columns=["CtN", "CtE", "CtRdRp", "PCR_Result"])
sangvi["PCR_Result"] = np.where(sangvi["PCR_Result"] > 0, "Positive", "Negative")
sangvi["CtN"] = sangvi["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtN that is correlated with CtE
sangvi["CtRdRp"] = 0.8 * sangvi["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtRdRp that is highly correlated with CtE
sangvi.to_csv("sangvi.csv", index=False)

# Generate random data for the Baner dataset
baner = pd.DataFrame(np.random.randn(100, 4), columns=["CtN", "CtE", "CtRdRp", "PCR_Result"])
baner["PCR_Result"] = np.where(baner["PCR_Result"] > 0, "Positive", "Negative")
baner["CtN"] = baner["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtN that is correlated with CtE
baner["CtRdRp"] = 0.8 * baner["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtRdRp that is highly correlated with CtE
baner.to_csv("baner.csv", index=False)

# Load the datasets from the CSV files
sangvi = pd.read_csv("sangvi.csv")
baner = pd.read_csv("baner.csv")

# Calculate the correlation matrix between the two datasets
corr = pd.concat([sangvi, baner]).corr()

# Plot the correlation heatmap
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"orientation": "horizontal", "ticks": [-1, -0.5, 0, 0.5, 1]}, vmin=-1, vmax=1, center=0, ax=ax)
plt.title("Correlation Heatmap between Sangvi and Baner Datasets")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data for the Sangvi dataset
sangvi = pd.DataFrame(np.random.randn(100, 4), columns=["CtN", "CtE", "CtRdRp", "PCR_Result"])
sangvi["PCR_Result"] = np.where(sangvi["PCR_Result"] > 0, "Positive", "Negative")
sangvi["CtN"] = sangvi["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtN that is correlated with CtE
sangvi["CtRdRp"] = 0.5 * sangvi["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtRdRp that is moderately correlated with CtE
sangvi.to_csv("sangvi.csv", index=False)

# Generate random data for the Baner dataset
baner = pd.DataFrame(np.random.randn(100, 4), columns=["CtN", "CtE", "CtRdRp", "PCR_Result"])
baner["PCR_Result"] = np.where(baner["PCR_Result"] > 0, "Positive", "Negative")
baner["CtN"] = baner["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtN that is correlated with CtE
baner["CtRdRp"] = 0.4 * baner["CtE"] + 0.2 * np.random.randn(100)  # add noise to CtRdRp that is moderately correlated with CtE
baner.to_csv("baner.csv", index=False)

# Load the datasets from the CSV files
sangvi = pd.read_csv("sangvi.csv")
baner = pd.read_csv("baner.csv")

# Calculate the correlation matrix between the two datasets
corr = pd.concat([sangvi, baner]).corr()

# Select only correlation values between 0.35 and 0.56
corr_filtered = corr[(corr >= 0.35) & (corr <= 0.56)]

# Plot the correlation heatmap
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corr_filtered, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"orientation": "horizontal", "ticks": [0.35, 0.4, 0.45, 0.5, 0.55, 0.56]}, vmin=0.35, vmax=0.56, center=0.45, ax=ax)
plt.title("Correlation Heatmap between Sangvi and Baner Datasets (Filtered)")
plt.show()
```
