```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Define the data set
date = df["Date"]
ctRdRp= df["CtRdRp"]

# Convert date strings to numeric values
date_num = np.arange(len(date))

# Reshape arrays to make them compatible with linear regression
date_num = date_num.reshape(-1, 1)
ctRdRp = np.array(ctRdRp).reshape(-1, 1)

# Create a linear regression model and fit it to the data
model = LinearRegression()
model.fit(date_num, ctRdRp)

# Print the coefficient
print('Linear regression coefficient: ', model.coef_[0][0])
```
