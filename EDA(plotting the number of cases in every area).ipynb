```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('data1.csv')

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Group the data by Date and count the number of Positive cases
daily_cases = df[df['PCR_Result'] == 'Positive'].groupby('Date').count()['PCR_Result']
# Group the data by Location and count the number of Positive cases
location_cases = df[df['PCR_Result'] == 'Positive'].groupby('Location').count()['PCR_Result']

# Plot the location-wise positive cases
plt.figure(figsize=(10, 6))
sns.barplot(x=location_cases.index, y=location_cases)
plt.xticks(rotation=90)
plt.title('Location-wise Positive Cases')
plt.xlabel('Location')
plt.ylabel('Number of Cases')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('data1.csv')

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Filter the data for years 2021, 2022, and 2023
df = df[(df['Date'].dt.year == 2021) | (df['Date'].dt.year == 2022) | (df['Date'].dt.year == 2023)]

# Group the data by Date and count the number of Positive cases
daily_cases = df[df['PCR_Result'] == 'Positive'].groupby('Date').count()['PCR_Result']

# Plot the daily positive cases trend
plt.figure(figsize=(10, 6))
plt.plot(daily_cases)
plt.title('Daily Positive Cases Trend for Years 2021, 2022, and 2023')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.show()
```
